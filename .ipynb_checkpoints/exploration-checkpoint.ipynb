# Preprocessing

I found that preprocessing enhances the performance -- including, grayscaling, cropping. 
Also we can try few-shot learning with some data I found.

```python
def crop_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert the image to grayscale
    blurred = cv2.GaussianBlur(gray, (35, 35), 0)  # Apply Gaussian blur to the image to reduce noise

    alpha = 1.9  # Contrast control (1.0-3.0)
    beta = 20  # Brightness control (0-100)
    contrasted = cv2.convertScaleAbs(blurred, alpha=alpha, beta=beta)

    edges = cv2.Canny(contrasted, 29, 29)  # Use Canny edge detection to find edges in the image

    lines = cv2.HoughLinesP(edges, 1, np.pi/180, 500, minLineLength=2000, maxLineGap=500)  # Use Hough Line Transform to find lines in the image
    
    min_x, min_y = image.shape[1], image.shape[0]
    max_x, max_y = 0, 0

    try:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Update the smallest and largest x and y values
            min_x, min_y = min(min_x, x1, x2), min(min_y, y1, y2)
            max_x, max_y = max(max_x, x1, x2), max(max_y, y1, y2)

        # Crop the image vertically and horizontally
        cropped = image[min_y:max_y, min_x:max_x]
        
        height, width = cropped.shape[:2]
        
        # Crop the image into halves
        middle_line_x = width // 2
        
        left_half = cropped[:, :middle_line_x]
        right_half = cropped[:, middle_line_x:]
        
        # Save each half as a separate file
        cv2.imwrite('left_half.jpg', left_half)
        cv2.imwrite('right_half.jpg', right_half)
        
        print("[INFO] Crop and splitting done!")
        return cropped
    
    except Exception as e:
        print(f"[INFO] No crop possible! Error: {str(e)}")
        return image
```

```python
import base64
import requests

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
def save_text(text, output_path):
    with open(output_path.replace('.jpg', '.txt'), 'w', encoding='utf-8') as f:
        f.write(text)
```

```python
base64_image = encode_image(path+'/right_bin.jpg')

# prompt = "This is a french document about declaration of succession. Recreate the table with the texts in the image and create a txt file:"
# prompt = "This is a french document about declaration of succession. Recreate the table with the texts in the image in a txt format:" #It's important that you are extra careful with numbers and names 
prompt = "This is a french document about declaration of succession. Recreate the table with the texts in the image. It's important that you are extra careful with numbers and names :" 

# prompt = "This is a frech document about declaration of succession. Recreate the table with the texts in the image:"
# prompt =  "This is a page of the declaration of succession of early 20th century Belgium in French. This page should have information about the name of the deseased, the registered date and the location of death, what assets and liability this person has left to the sucessors. Recreate the table with the texts in the image:"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}    

payload = {
    "model": "gpt-4o",
    "messages": [
    {
        "role": "user",
        "content": [
        {
            "type": "text",
            "text": prompt
        },
        {
            "type": "image_url",
            "image_url": {
            "url": f"data:image/jpeg;base64,{base64_image}"
            }
        }
        ]
    }
    ],
    "max_tokens": 1000,
    "temperature": 0
}

response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
```
